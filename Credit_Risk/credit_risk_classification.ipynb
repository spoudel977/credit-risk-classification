```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
```

---

## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame

# Read the CSV file
lending_data = pd.read_csv('Resources/lending_data.csv')

# Display the first few rows of the dataframe
lending_data.head()
```

```python
# Remove rows with missing values
lending_data_cleaned = lending_data.dropna()
```

```python
# Review the DataFrame
lending_data_cleaned.describe()
```

### Step 2: Create the labels set (`y`)  from the “loan_status” column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features

# Separate the y variable, the labels
y = lending_data_cleaned["loan_status"]

# Separate the X variable, the features
X = lending_data_cleaned.drop(columns=["loan_status"])
```

```python
# Review the y variable Series
y.head()
```

```python
# Review the X variable DataFrame
X.head() 
```

```python
# Review the X variable DataFrame
X.info()
```

```python
# Review the X variable DataFrame
X.describe()
```

### Step 3: Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split


# Split the data into training and testing sets with random_state=1
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    test_size=0.2, 
                                                    random_state=1)
```

```python
# Display the shapes of the resulting sets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

---

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the logistic regression model with random_state=1
model = LogisticRegression(solver='lbfgs', random_state=1)

# Train / Fit the model using the training data
model.fit(X_train, y_train)
```

# Validate the model using the test data

```python
# Score the model
print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python
# Make a prediction using the testing data
y_pred = model.predict(X_test)
pd.DataFrame({"Prediction": y_pred, "Actual": y_test})
```

```python
# Display the first few predictions
print(y_pred[:10])
```

### Step 3: Evaluate the model’s performance by doing the following:

* Generate a confusion matrix.

* Print the classification report.

```python
# Generate a confusion matrix for the model
cm = confusion_matrix(y_test, y_pred)
```

```python
# Print the classification report for the model
# Generate the classification report
report = classification_report(y_test, y_pred)

# Print the classification report
print(report)
```
